```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from ontocluster import OntoClusterer
```

```python
data = pd.read_csv('HRdata.csv', index_col = 'Unnamed: 0')
X = pd.get_dummies(data)
X = X.astype(int)

pca = PCA()
X_dec = pca.fit_transform(X)
```

```python
plt.scatter(X_dec[:, 0], X_dec[:, 1]);
```

```python
model = OntoClusterer('hdbscan')
model.score(X)
```

```python
model = OntoClusterer('xmeans')
onto = model.fit(X, data)
onto.save('HRdata.owl')
```

```python
plt.scatter(X_dec[:, 0], X_dec[:, 1], c = model.labels);
```
